seaborn->Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures.
Seaborn helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. Its dataset-oriented, declarative API lets you focus on what the different elements of your plots mean, rather than on the details of how to draw them.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

load_dataset->Load an example dataset from the online repository (requires internet).
This function provides quick access to a small number of example datasets that are useful for documenting seaborn or generating reproducible examples for bug reports. It is not necessary for normal usage.

```python
df=sns.load_dataset('tips')#for getting new dataset
df.head()
```

```python
df.columns=['total_bill', 'tip', 'gender', 'smoker', 'day', 'time', 'member']
```

```python
df
```

scatterplot->gives the scatterplot<br>arguments->xaxis,yaxis an data for reference

```python
# sns.lineplot()
# sns.boxplot()
# sns.scatterplot()
# sns.countplot()
# sns.barplot()
# sns.histplot()
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.show()
```

relplot=relational plot for numerical<br>
catplot=categorical plot for one numerical and another categorical

argumnets->hue->It varies the colors in a plot based on a categorical variable.,style->It changes the markers or line styles in a plot.

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',style='smoker')
```

col->gives two graphs on the basis of column provided

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',style='smoker',col='time')
plt.show()
```

```python
female=df[df['gender']=='Female']
```

```python
sns.relplot(x='tip',y='total_bill',data=female,hue='smoker',col='smoker')
```

```python
sns.relplot(x='tip',y='total_bill',data=female,hue='smoker',col='day')
```

row->for giving different rows on basis of column

```python
sns.relplot(x='tip',y='total_bill',data=female,hue='smoker',row='day')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',col='day')
plt.show()
```

kind->for getting line plot

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='line')#scatter plot
```

univariate analysis<br>
bivariate analysis<br>
multivariate analysis<br>

```python
sns.relplot(x='tip',y='total_bill',data=female,hue='smoker',row='day')
plt.show()
```

countplot->shows the counts of observation

```python
sns.countplot(df['day'])
```

catplot->for visualizing categorical data

```python
sns.catplot(x='day',y='total_bill',data=df,kind='violin',)
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='boxen',)
```
