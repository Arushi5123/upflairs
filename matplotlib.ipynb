```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('Used_Bikes.csv')
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
group=df.groupby('brand')
group
```

```python
price_of_ktm=list(group['price'].get_group('KTM'))
age=list(group['owner'].get_group('KTM'))
```

```python
plt.plot(price_of_ktm,age,marker="*",ms='12',mfc='plum',color='orchid')
```

```python
x=list(df['owner'].value_counts().keys())
y=list(df['owner'].value_counts().values)
plt.figure(figsize=(5,5))
plt.bar(x,y,color='violet',ls='dashed',ec='r',lw=0.9,width=0.2)
plt.xlabel('KEYS')
plt.ylabel('VALUES')
plt.title('OWNER')
plt.xticks(rotation=45)
plt.show()
```
