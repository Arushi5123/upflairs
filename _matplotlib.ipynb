<h>Matplotlib</h> 
<p>Matplotlib is a library in python used for data visualization by representing the data in the form of graphs.
As we know that knowing the relationship between two attributes can't be done easily by tabular form of data, this is where matplotlib comes in, which help to convert the tabular data into graphs or chart </p> 

<p>matplotlib have various functions to perform some tasks, some of them are as follows with some examples</p>
<h>plot</h>
<p>plot is a funtion used to plot a line graph between x and y axis<br>
This function takes various arguments like list ,marker, markersize(ms), marker face color(mfc) and color. It returns a line plot</p>

```python
ls=[3,5,7,5,8,2]
#linePlot()
plt.plot(ls,marker="*",ms=12,mfc='turquoise')#ms= marker size mfc=marker face color
```

```python
x=[3,5,7,5,8,2]
y=[6,4,6,3,8,3]
plt.plot(x,y,marker="*",ms=12,mfc='brown')
```

```python
x=[1,4,9,16]
y=[1,2,3,4]
p=plt.plot(x,y,marker="*",ms=12,mfc='k')# k=black
plt.show()
```

<h>Scatter, xlabel, ylabel, title</h>
<p>scatter=used when we want the relational distribution<br>xlabel=for labeling the x axis<br>ylabel=for labeling the y axis<br>title=for labeling the title</p>



```python
price=list(df['price'].values)
age=list(df['age'].values)
#2 columns must be numerical
plt.scatter(age,price,color='turquoise',marker='*',)
plt.xlabel("AGE")
plt.ylabel("PRICE")
plt.title("AGE vs PRICE")
plt.show()
```

<h>Bar Plot,figure,xticks</h>
<p>*Barplot=when we want to show the distribution of data points or oerform a comparison of metric<br>*figure=to define the figure of bar by length and width<br>*xticks for rotating the keys on x axis so that they don't overlap</p>

```python
brands=list(df['brand'].value_counts().keys())
no_of_counts=list(df['brand'].value_counts().values)
plt.figure(figsize=(14,7))# width height
plt.bar(brands,no_of_counts,color='salmon',ec='k',lw=1,ls='dashed',width=0.9)#ec=edge color lw=line width ls=line style--dashed, dotted
plt.xlabel('BRANDS')
plt.ylabel('N0_OF _COUNTS')
plt.title("NO. OF BIKE IN BRANDS")
plt.xticks(rotation=120)
plt.show()
#barplot
```

<h>Pie plot</h>
<p>pieplot=for graphing qualitative data or we need to know which of them chas more contribution in the events in form of percentage</p>

```python
plt.pie(no_of_counts,labels=ownership,autopct='%1.2f%%')
plt.figure(figsize=(8,8))
plt.show()
```

<p>savefig=used for saving the figure</p>

```python
country=['india','china','japan','pakistan']
population=[900,700,500,350]
explosion=[0,0.2,0,0.2]
plt.figure(figsize=((8,8)))
plt.pie(population,labels=country,autopct='%1.2f%%',wedgeprops={'ec':'k'},explode=explosion,shadow=True)
plt.savefig('my_pie_plot')#for saving the figure
plt.show()
```

<h>Boxplot</h>
<p>boxplt=it is mainly used when we want to know the outliers(exceptions) in the data which the machine learning algorithms can't comprehend</p> 

<p>In the box shown below the middle line is median the line above is 75 and the line bellow it is 25%<br>The dots seen at the top of the figure represents the outliers<br>The line which reachs the top represents maximum and the line reaches the bottom is the minimum</p>

```python
#box plot
plt.boxplot(bullet['price'])
plt.show()
```

<h>percentile formula=index+(per/100)*(n+1)</h>

```python
bullet_filter=bullet[(bullet['price']>=5000)&(bullet['price']<=200000)]
bullet_filter
```

```python
plt.boxplot(bullet_filter['price'])
plt.show()
```
