```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Used_Bikes.csv')
df.head()
```

```python
df.info()
#regression
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['brand'].value_counts().keys()
brand_list=['Bajaj', 'Royal Enfield', 'Hero', 'Honda', 'Yamaha', 'TVS', 'KTM',
       'Suzuki', 'Harley-Davidson', 'Kawasaki', 'Hyosung', 'Mahindra',
       'Benelli', 'Triumph', 'Ducati', 'BMW']
```

```python
df=df[df['brand'].isin(brand_list)]
```

```python
categorical=df.select_dtypes(include='O')
categorical.head()
```

```python
numerical=df.select_dtypes(exclude='O')
numerical.head()
```

```python
label={'First Owner':1,
       'Second Owner' :2,         
        'Third Owner':3,            
       'Fourth Owner Or More':4 }
```

```python
categorical['owner'].value_counts()
```

```python
categorical['owner']=categorical['owner'].map(label)
```

```python
#brand_list
dt={}
for i in range(0,len(brand_list)):
    dt[brand_list[i]]=i

```

```python
dt
```

```python
categorical['brand']=categorical['brand'].map(dt)#can't run two times
```

```python
categorical
```

```python
categorical['owner_name']='Arushi Mathur'
categorical
```

```python
categorical.drop(['bike_name','city','owner_name'],axis=1,inplace=True)
```

```python
df2=pd.concat([categorical,numerical],axis=1)
df2
```

here x and y are the features and target respectively,then we spilt the two to x_train,x_test for x and y _train and y_test for y.<br>Now the x_train and y_train are used for training the machine learning algorithm and x_test,y_test are used for testing the algorithm

```python
x=df2.drop('price',axis=1)
y=df[['price']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

train_test_split->Split arrays or matrices into random train and test subsets.

```python
#x and y-->training and testing split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2) #test size=20% other remining training ==80%
#we can also define train_size
```

```python
#the prediction given by algorithm is called y hat
#the x_test is given as input and the y_test is the actual value
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dfc=DecisionTreeRegressor()
```

```python
dfc.fit(x_train,y_train)
```

```python
print(dfc.score(x_train,y_train))
print(dfc.score(x_test,y_test))
```

```python
pred=dfc.predict(x_test)
```

```python
import joblib
```

```python
joblib.dump(dfc,'decisiontreeregressor.lb')
```

```python
from sklearn.linear_model import LinearRegression# LinearRegression is a class
```

```python
lg=LinearRegression()#crating object of class LinearRegression
```

```python
#Algorithm training 
lg.fit(x_train,y_train)#training for the training purpose
```

```python
#model evaluation or testing the model performance
#both training and testing
print("Training score:",round(lg.score(x_train,y_train)*100,2))
print("Testing score:",round(lg.score(x_test,y_test)*100,2))
#>90
#difference between training score and test score should be >5
#model overfit or underfit pr perfect(generailized model)

#model overfit 
#Training score--->high
#Testing score--->low

#model underfit or under
#Training score--->low
#Testing score--->high or low

#generalized model
#difference not more than 5
#no model overfit or underfit 
#score or accuracy high in both case on train as well as test data
```

```python
#prediction
y_hat=lg.predict(x_test)#1462
y_hat
```

```python
x_test
```

```python
y_test['prediction']=y_hat #actual price
```

```python
y_test['prediction']=y_test['prediction'].map(lambda x:round(x,2))
```

```python
print("Training score:",round(rdf.score(x_train,y_train)*100,2))
print("Training score:",round(rdf.score(x_test,y_test['price'])*100,2))
```
