```python
import pandas as pd
```

```python
df=pd.read_csv('Used_Bikes.csv')
```

```python
df
```

```python
df.drop(['bike_name','city'],axis=1,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
categorical=df.select_dtypes(include='O')
categorical
```

```python
numerical=df.select_dtypes(exclude='O')
numerical
```

```python
categorical['owner'].value_counts()
```

```python
label={'First Owner':1,
    'Second Owner':2,
    'Third Owner':3,
    'Fourth Owner Or More':4}
label
```

```python
categorical['owner']=categorical['owner'].map(label)
```

```python
categorical
```

```python
df['brand'].value_counts().keys()
arr=['Bajaj', 'Royal Enfield', 'Hero', 'Honda', 'Yamaha', 'TVS', 'KTM',
       'Suzuki', 'Harley-Davidson', 'Kawasaki', 'Hyosung', 'Mahindra',
       'Benelli', 'Triumph', 'Ducati', 'BMW', 'Jawa', 'Indian', 'MV',
       'Rajdoot', 'LML', 'Yezdi', 'Ideal']
```

```python
dt={}
for i in range(0,len(arr)):
    dt[arr[i]]=i
dt
```

```python
categorical['brand']=categorical['brand'].map(dt)
```

```python
categorical
```

```python
df2=pd.concat([categorical,numerical],axis=1)
df2
```

```python
x=df2.drop('price',axis=1)
y=df2[['price']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression  
```

```python
lg=LinearRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
print('training score',round(lg.score(x_train,y_train)*100,2))
print('testing score',round(lg.score(x_test,y_test)*100,2))
```

```python
y_hat=lg.predict(x_test)
```

```python
y_test['prediction']=y_hat
```

```python
y_test['prediction']=y_test['prediction'].map(lambda x:round(x,2))
```
